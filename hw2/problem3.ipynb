```python
def get_ic(text):
    text = text.upper().replace(" ", "")
    total = len(text)
    frequencies = {chr(i): 0 for i in range(ord('A'), ord('Z') + 1)}
    for char in text:
        frequencies[char] += 1
    if total == 0:
        return 0
    ic = sum(frequency * (frequency - 1) for frequency in frequencies.values()) / (total * (total - 1))
    return ic

ciphertext = "TTEUM GQNDV EOIOL EDIRE MQTGS DAFDR CDYOX IZGZP PTAAI TUCSI XFBXY SUNFE SQRHI SAFHR TQRVS VQNBE EEAQG IBHDV SNARI DANSL EXESX EDSNJ AWEXA ODDHX EYPKS YEAES RYOET OXYZP PTAAI TUCRY BETHX UFINR"
for length in range(2, 11):
    ic = 0
    for i in range(length):
        partial_text = ciphertext[i::length]
        ic += get_ic(partial_text)
    avg_ic = ic / length
    print(f"Key length: {length}, Average IC: {avg_ic}")
```

```python
def caesar_decrypt(ciphertext, shift):
    decrypted = ""
    for char in ciphertext:
        if char.isalpha():
            shifted = ord(char) - shift
            if char.islower():
                if shifted < ord('a'):
                    shifted += 26
            elif char.isupper():
                if shifted < ord('A'):
                    shifted += 26
            decrypted += chr(shifted)
        else:
            decrypted += char
    return decrypted

for i in range(5):
    group = "TTEUM GQNDV EOIOL EDIRE MQTGS DAFDR CDYOX IZGZP PTAAI TUCSI XFBXY SUNFE SQRHI SAFHR TQRVS VQNBE EEAQG IBHDV SNARI DANSL EXESX EDSNJ AWEXA ODDHX EYPKS YEAES RYOET OXYZP PTAAI TUCRY BETHX UFINR"[i::5]
    print(f"Group {i+1}: {group}")
    for shift in range(26):
        decrypted_group = caesar_decrypt(group, shift)
        print(f"Shift {shift}: {decrypted_group}")
```
